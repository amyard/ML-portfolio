```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = 'https://www.weblancer.net/jobs/?type=project'
main_url = 'https://www.weblancer.net'

def get_html(url):
    response = urllib.request.urlopen(url)
    return response.read()

def get_page_count(html):
    soup = BeautifulSoup(html,'lxml')
    footer = soup.find('div', class_='pagination_box').find_all('a')[-2].text
    return int(footer)



def parse(html):
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('div', class_='page_content')

    projects = []
    for row in table.find_all('div', class_='row')[:-1]:
        projects.append({
            'title': row.a.text,
            'category2':[category.text for category in row.find('div', class_='col-sm-8').find_all('a')],
            'price2': [price.text.strip() for price in row.find('div', class_='col-sm-2').find_all('div', class_='title')],
            'zajavok2': [app.text.strip().split(' ')[0] for app in row.find('div', class_='col-sm-2').find_all('div', class_='text_field')],
            'url': main_url+row.find('h2').find('a').get('href')
            })

    return projects




page_count = get_page_count(get_html(BASE_URL))
print('Всего найдено страниц %d' % page_count)
    
projects = []
for page in range(1, page_count+1):
    print('Парсинг %d%%' %  (page/page_count*100))
    projects.extend(parse(get_html(BASE_URL+'&page=%d' %page)))
        
df = pd.DataFrame.from_dict(projects)
df['price'] = [i[0] for i in df['price2']]
df['zajavok'] = [i[0] for i in df['zajavok2']]
df['category'] = [i[0] for i in df['category2']]
df.drop(['price2','zajavok2','category2'], axis =1, inplace = True)
df.head(10)
```
