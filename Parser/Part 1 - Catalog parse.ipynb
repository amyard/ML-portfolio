```python
import requests
from bs4 import BeautifulSoup
import csv

import warnings
warnings.filterwarnings('ignore')


def get_html(url):
    response = requests.get(url)
    return response.text

def get_total_pages(html):
    soup = BeautifulSoup(html, 'lxml')
    total_pages = soup.find('div', class_='pager rel clr').find_all('span', class_='item fleft')[-1].text
    return int(total_pages)

def save(projects, path):
      with open(path,'w') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(('Заголовок',"Страница","Район", "Дата","Цена"))
            
            for project in projects:
                writer.writerow((project['title'], project['url'], project['district'], project['date'], project['price']))

def get_data(html):
    soup = BeautifulSoup(html)
    ads = soup.find_all('tr', class_='wrap')
    
    projects = []
    for ad in ads:
        projects.append({
            'title': ad.find('h3', class_='x-large').find('a').text.strip(),
            'url':ad.find('h3', class_='x-large').find('a').get('href'),
            'district': ad.find_all('div', class_='space rel')[1].find('span').text.strip().split(',')[1],
            'date':ad.find_all('div', class_='space rel')[1].find_all('p')[1].text.strip(),
            'price':ad.find_all('div', class_='space inlblk rel')[0].text.strip()
        })

    return projects

def main():
    url = 'https://www.olx.ua/nedvizhimost/kvartiry-komnaty/arenda-kvartir-komnat/odessa/?page=1'
    base_url = 'https://www.olx.ua/nedvizhimost/kvartiry-komnaty/arenda-kvartir-komnat/odessa/?page='
    
    total_pages = get_total_pages(get_html(url))
    print("Всего найдено страниц: %d " %total_pages)
    
    projects = []
    
    for page in range(1,total_pages+1):
        print("Парсинг %d%% " %(page/total_pages*100))
        
        projects.extend(get_data(get_html(base_url+str(page))))
    
    save(projects,'project.csv')
    
    
if __name__ == '__main__':
    main()
```
