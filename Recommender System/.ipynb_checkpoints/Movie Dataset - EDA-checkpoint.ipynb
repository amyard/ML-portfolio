## Movie Release Dates

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

def get_month(x):
    try:
        return month_order[int(str(x).split('-')[1]) - 1]
    except:
        return np.nan

def get_day(x):
    try:
        year, month, day = (int(i) for i in x.split('-'))    
        answer = datetime.date(year, month, day).weekday()
        return day_order[answer]
    except:
        return np.nan
```

```python
df['day'] = df['release_date'].apply(get_day)
df['month'] = df['release_date'].apply(get_month)
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x='month', data=df, order=month_order)
ax.set(title = "Number of Movies released in a particular month.")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+0.2,height+50, '%.0f'%(height))
plt.show()
```

```python
month_mean = pd.DataFrame(df[df['revenue'] > 1e8].groupby('month')['revenue'].mean())
month_mean['mon'] = month_mean.index

plt.figure(figsize=(15,6))
ax = sns.barplot(x='mon', y='revenue', data=month_mean, order=month_order)
ax.set(title = "Average Gross by the Month for Blockbuster Movies")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()-0.07,height+5000000, '{:,}'.format(round(height)))
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(15, 8))
sns.boxplot(x='month', y='return', data=df[df['return'].notnull()], palette="muted", ax =ax, order=month_order)
ax.set_ylim([0, 12])

# The months of June and July tend to yield the highest median returns. September is the least successful months on the aforementioned metrics.
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='day', data=df, order=day_order)
ax.set(title = 'Number of Movies released on a particular day.')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+0.2,height+50, '%.0f'%(height))
plt.show()
```

```python
credits_df = pd.read_csv('credits.csv')
credits_df.head()
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan

df['id'] = df['id'].apply(convert_int)
df[df['id'].isnull()]
```

```python
df = df.drop([19730, 29503, 35587])
df['id'] = df['id'].astype('int')
```

```python
df = df.merge(credits_df, on='id')
df.shape
```

```python
df['cast'] = df['cast'].apply(ast.literal_eval)
df['crew'] = df['crew'].apply(ast.literal_eval)

df['cast_size'] = df['cast'].apply(lambda x: len(x))
df['crew_size'] = df['crew'].apply(lambda x: len(x))

df['cast'] = df['cast'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```
